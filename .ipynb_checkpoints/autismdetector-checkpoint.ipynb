```python
# Step 1: Import required libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Step 2: Set paths and parameters
train_dir = "dataset/train"
valid_dir = "dataset/valid"

IMG_SIZE = (224, 224)   # VGG16 required size
BATCH_SIZE = 32
EPOCHS = 10
```

```python
# Step 3: Create data generators with augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

valid_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False  # important for evaluation
)
```

```python
# Step 4: Load VGG16 without top layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze all convolutional layers
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Step 5: Add custom classification layers
x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)  # binary classification

model = Model(inputs=base_model.input, outputs=output)
```

```python
# Step 6: Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# Step 7: Train the model
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=valid_generator
)

```

```python
# Step 8: Evaluate the model
loss, accuracy = model.evaluate(valid_generator)
print(f"Validation Loss: {loss:.4f}")
print(f"Validation Accuracy: {accuracy:.4f}")
```

```python
# Step 9: Get predictions for evaluation
y_pred = model.predict(valid_generator)
y_pred_classes = (y_pred > 0.5).astype(int)

print(classification_report(valid_generator.classes, y_pred_classes, target_names=['Autistic', 'Non_Autistic']))

cm = confusion_matrix(valid_generator.classes, y_pred_classes)
print("Confusion Matrix:\n", cm)
```

```python
model.save("autism_detection_vgg16.h5")
print("autism_detection_vgg16.h5")
```

### Hyperparameter tuning 

```python
import tensorflow as tf

# Load trained model
model = tf.keras.models.load_model("autism_detection_vgg16.h5")

# Unfreeze layers whose name starts with 'block5'
for layer in model.layers:
    if layer.name.startswith("block5"):
        layer.trainable = True
    else:
        layer.trainable = False  # Keep others frozen

print("Trainable status of layers:")
for layer in model.layers:
    print(f"{layer.name}: {layer.trainable}")
```

```python
from tensorflow.keras.optimizers import Adam

# Recompile with a lower learning rate to avoid damaging pretrained weights
model.compile(
    optimizer=Adam(learning_rate=1e-5),  # Fine-tuning uses very small LR
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing import image
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# --- Part 1: Activity Recommendation Engine ---

def create_activity_database():
    """Creates a sample DataFrame of activities. In a real application,
    this would be loaded from a CSV file or database."""
    data = {
        'activity_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
        'name': [
            "Sensory Rice Bin", "Play-Doh Sculpting", "Swinging", "Threading Beads",
            "Simple Board Game", "Social Story Time", "Bubble Blowing", "Trampoline Jumping",
            "Picture Exchange (PECS)", "Building with LEGOs"
        ],
        'skills_targeted': [
            "sensory_integration fine_motor calming",
            "fine_motor sensory_integration creative",
            "sensory_integration gross_motor vestibular calming",
            "fine_motor concentration hand_eye_coordination",
            "social_skills turn_taking rules_following",
            "social_skills communication comprehension emotional_regulation",
            "communication oral_motor joint_attention",
            "gross_motor sensory_integration proprioceptive energy_release",
            "communication non_verbal requesting choice_making",
            "fine_motor creative problem_solving concentration"
        ],
        'age_range': ["3-5", "3-8", "3-10", "4-7", "5-10", "4-10", "3-6", "4-10", "3-8", "5-10"]
    }
    return pd.DataFrame(data)

class ActivityRecommender:
    """A data-driven recommender for autism support activities."""
    def __init__(self, activities_df):
        self.activities_df = activities_df
        self.tfidf_vectorizer = TfidfVectorizer(stop_words='english')
        self.tfidf_matrix = self.tfidf_vectorizer.fit_transform(self.activities_df['skills_targeted'])

    def recommend(self, child_profile_needs, num_recommendations=3):
        """Recommends activities based on a child's profile needs."""
        profile_vector = self.tfidf_vectorizer.transform([child_profile_needs])
        cosine_similarities = cosine_similarity(profile_vector, self.tfidf_matrix).flatten()
        related_activities_indices = cosine_similarities.argsort()[::-1]
        top_indices = [idx for idx in related_activities_indices if cosine_similarities[idx] > 0][:num_recommendations]
        
        if not top_indices:
            return pd.DataFrame()
        return self.activities_df.iloc[top_indices]

# --- Part 2: Image Prediction Module ---

def load_prediction_model(model_path="autism_detection_vgg16_finetuned.h5"):
    """Loads the trained Keras model from disk."""
    try:
        model = tf.keras.models.load_model(model_path)
        print(f"Successfully loaded model '{model_path}'")
        return model
    except Exception as e:
        print(f"Error loading model: {e}")
        print(f"Please ensure the model file '{model_path}' is in the same directory.")
        return None

def predict_from_image(image_path, model):
    """Loads an image, preprocesses it, and predicts using the loaded Keras model."""
    if model is None:
        return "Error: Model is not loaded."
    try:
        IMG_SIZE = (224, 224)
        img = image.load_img(image_path, target_size=IMG_SIZE)
        img_array = image.img_to_array(img)
        img_array /= 255.0
        img_batch = np.expand_dims(img_array, axis=0)
        prediction_value = model.predict(img_batch)[0][0]
        
        print(f"Analyzing '{image_path}'... Raw prediction score: {prediction_value:.4f}")

        if prediction_value > 0.5:
            return "Non_Autistic"
        else:
            return "Autistic"
    except FileNotFoundError:
        return f"Error: The file was not found at '{image_path}'"
    except Exception as e:
        return f"An error occurred during prediction: {e}"

# --- Part 3: The Main Workflow Integration ---

def get_human_input_from_ui_checklist(selected_options):
    """Simulates receiving input from a UI checklist."""
    print("\nReceived user input from checklist...")
    return " ".join(selected_options)

def main_workflow(image_path, user_selected_needs, model, recommender):
    """Orchestrates the entire process from prediction to recommendation."""
    print("--- Starting Autism Support Workflow ---")
    
    # Step 1: Get prediction from the vision model
    prediction = predict_from_image(image_path, model)
    print(f"Facial Analysis Prediction: {prediction}")
    
    # Step 2: If prediction is "Autistic", ask for human input and proceed
    if prediction == "Autistic":
        print("\nPrediction suggests potential ASD. Proceeding to gather specific needs...")
        
        child_needs_string = get_human_input_from_ui_checklist(user_selected_needs)
        print(f"Child's needs profile: '{child_needs_string}'")
        
        # Step 3: Get recommendations based on the human input
        print("\n--- Generating Personalized Activity Recommendations ---")
        recommendations = recommender.recommend(child_needs_string, num_recommendations=3)
        
        if not recommendations.empty:
            print(recommendations.to_string(index=False))
        else:
            print("Could not find matching activities for the specified needs.")
    else:
        print("\nWorkflow complete. No specific activities recommended based on the prediction.")
    
    print("\n--- End of Workflow ---")

# --- Example Execution ---

if __name__ == "__main__":
    # 1. Load the ML model and initialize the recommender ONCE
    prediction_model = load_prediction_model()
    activity_db = create_activity_database()
    activity_recommender = ActivityRecommender(activity_db)

    # Only proceed if the model was loaded successfully
    if prediction_model:
        # 2. SIMULATE a user uploading an image and filling out a checklist
        # IMPORTANT: Replace this with the actual path to a test image
        test_image = "dataset/test/autistic/Autistic.0 11.39.10.jpg"
        
        # This list simulates the user ticking checkboxes in a UI
        user_choices = ["fine_motor", "calming", "sensory_integration"]

        # 3. Run the complete, integrated workflow
        main_workflow(test_image, user_choices, prediction_model, activity_recommender)
```
